matplotlib:- it is useful for making plot.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel('angle')
plt.ylabel('value')
plt.title('parabola curve')
plt.plot(x,y)
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
language =['eng','hindi','sanskrit','french','spanish']
people=[10,145,15,5,56]
plt.xlabel('language')
plt.ylabel('number of people speaking')
ax.bar(language,people)
plt.show()
```

```python
figl=plt.figure()
ax=figl.add_axes([0,0,1,1])
language =['english','spanish','hindi','french','latin']
people=[100,25,200,155,167]
ax.pie(people,labels=language,autopct='%1.1f%%')
plt.show()
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='g')
plt.show()
```

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
